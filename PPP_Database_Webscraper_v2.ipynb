```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import urllib.request
import bs4 as bs
import re
```

```python
num_comp_per_page = 200
num_pages = 24447
```

```python
df_pg1 = pd.read_html("https://ppp.directory/search?business_name=&loan_category%5B%5D=h&loan_category%5B%5D=g&loan_category%5B%5D=f&loan_category%5B%5D=e&loan_category%5B%5D=d&loan_category%5B%5D=c&loan_category%5B%5D=b&loan_category%5B%5D=a&city=&county=&zip=&naics_code=&lender=")
```

```python
## Combine all companies from the first page into a single dataframe of structured data ##


df_list_pg1 = []

for x in range(1, num_comp_per_page + 1):
    
    df_pg1_tr = df_pg1[x].T
    
    ## slice off the first row and 8 variables into a single data frame ##
    df_pg1_tr1 = df_pg1_tr.iloc[:2, :]
    df_pg1_tr1 = df_pg1_tr1.rename(columns=df_pg1_tr1.iloc[0])
    df_pg1_tr1.reset_index(inplace=True)
    df_pg1_tr1 = df_pg1_tr1.iloc[1:, 1:]

    ## slice off the third row and next 8 variables into a single data frame ##
    df_pg1_tr2 = df_pg1_tr.iloc[2:, :]
    df_pg1_tr2 = df_pg1_tr2.rename(columns=df_pg1_tr2.iloc[0])
    df_pg1_tr2.reset_index(inplace=True)
    df_pg1_tr2 = df_pg1_tr2.iloc[1:, 1:]
    df_pg1_tr2

    ## combine the two dataframes into a single df with 16 variables ##
    ## now 1 row = 1 company ##

    df_concat = pd.concat([df_pg1_tr1, df_pg1_tr2], axis=1)
    df_list_pg1.append(df_concat)

df_pg1_final = pd.concat(df_list_pg1)
df_pg1_final.reset_index(inplace=True)
df_pg1_final = df_pg1_final.iloc[:, 1:]
df_pg1_final.rename(columns={'City, State Zip':'City_State_Zip'}, inplace=True)
df_pg1_final['City'] = df_pg1_final.City_State_Zip.str.slice(stop=-10)
df_pg1_final['State'] = df_pg1_final.City_State_Zip.str.slice(start=-8, stop=-6)
df_pg1_final['Zip'] = df_pg1_final.City_State_Zip.str.slice(start=-5)
df_pg1_final
```

```python
df_pg1_final = df_pg1_final[["Business Name", "State", "City", "Zip", "Race or Ethnicity", "Gender", "Veteran"\
                     , "Non-Profit", "Loan Amount", "Jobs Retained", "Date Approved", "NAICS Code",\
                    "NAICS Description", "NAICS Category", "Lender", "Congressional District"]]

df_pg1_final
```

```python
## Create a dictionary of all page URL's from page 2: num_pages ##

#empty dictionary
ppp_pages = {} 

# for loop to create pagex = urlx for all pages of PPP database #
for x in range(2, num_pages+1): 
    ppp_pages["{0}".format(x)]="https://ppp.directory/search?page=" + str(x) + "&business_name=&loan_category%5B0%5D=h&loan_category%5B1%5D=g&loan_category%5B2%5D=f&loan_category%5B3%5D=e&loan_category%5B4%5D=d&loan_category%5B5%5D=c&loan_category%5B6%5D=b&loan_category%5B7%5D=a&city=&county=&zip=&naics_code=&lender="
```

```python
ppp_page_list = []
for value in ppp_pages.values():
    ppp_page_list.append(value)
```

```python
## Create List of Data Frames for Pages 2 to num_pages ##

df_pages_list = []

for url in ppp_page_list:
    df_url = pd.read_html(url)
    df_pages_list.append(df_url)
```

```python
df_list_4 = []

for df in df_pages_list:
    for x in range(1, num_comp_per_page + 1):
    
        df_html_tr = df[x].T

        ## slice off the first row and 8 variables into a single data frame ##
        df_html_tr1 = df_html_tr.iloc[:2, :]
        df_html_tr1 = df_html_tr1.rename(columns=df_html_tr1.iloc[0])
        df_html_tr1.reset_index(inplace=True)
        df_html_tr1 = df_html_tr1.iloc[1:, 1:]

        ## slice off the third row and next 8 variables into a single data frame ##
        df_html_tr2 = df_html_tr.iloc[2:, :]
        df_html_tr2 = df_html_tr2.rename(columns=df_html_tr2.iloc[0])
        df_html_tr2.reset_index(inplace=True)
        df_html_tr2 = df_html_tr2.iloc[1:, 1:]
        df_html_tr2

        ## combine the two dataframes into a single df with 16 variables ##
        ## now 1 row = 1 company ##

        df_concat = pd.concat([df_html_tr1, df_html_tr2], axis=1)
        df_list_4.append(df_concat)

df_all = pd.concat(df_list_4)
df_all.reset_index(inplace=True)
df_all = df_all.iloc[:, 1:]
df_all.rename(columns={'City, State Zip':'City_State_Zip'}, inplace=True)
df_all['City'] = df_all.City_State_Zip.str.slice(stop=-10)
df_all['State'] = df_all.City_State_Zip.str.slice(start=-8, stop=-6)
df_all['Zip'] = df_all.City_State_Zip.str.slice(start=-5)
```

```python
df_all = df_all[["Business Name", "State", "City", "Zip", "Race or Ethnicity", "Gender", "Veteran"\
                     , "Non-Profit", "Loan Amount", "Jobs Retained", "Date Approved", "NAICS Code",\
                    "NAICS Description", "NAICS Category", "Lender", "Congressional District"]]

df_all
```

```python
## Add in the Page 1 Dataframe to create one combined Dataframe ##

df_combined = pd.concat([df_pg1_final, df_all], ignore_index = True, sort = False)

df_combined["Jobs Retained"] = df_combined["Jobs Retained"].apply(pd.to_numeric)

print(df_combined.dtypes)
```
